# histogram


```python
import matplotlib.pyplot as plt
```

```python
blood_sugar=[113,84,23,56,6,45,67,234,345,322,45,57,12]
plt.hist(blood_sugar,rwidth=0.8)
```

# bins parameter

```python
plt.hist(blood_sugar,rwidth=0.5,bins=4)
```

```python
plt.xlabel('sugar levels')
plt.ylabel('number of patients')
plt.title('blood sugar chart')
plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.95,color='b')
```

```python
plt.xlabel('sugar levels')
plt.ylabel('number of patients')
plt.title('blood sugar chart')

blood_sugar_men=[113,234,453,56,32,435,56,789,45,65,67,12,54]
blood_sugar_women=[19,98,87,65,54,32,123,342,546,532,34,223,45]
plt.hist(blood_sugar_men,blood_sugar_women,bins=[80,100,125,150],rwidth=0.95,color=['b','g'])
plt.legend()
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['green','orange'],label=['men','women'])
plt.legend()
```

# histtype=step

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")
plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.95,histtype='step')
```
