# cleaning data
cleaning data is a crucial step before EDA as it will remove the ambigous data that can affect the outcome of EDA.

while cleaning data we can perform foloowing steps\

1.Remove duplicate rows

2.Handling missing values

3.converting columns to appropriate datatype

4.adding important columns

# step one removing duplicate rows if any

```python
df1[df1.duplicated()].shape
df1.duplicated().sum()
```

```python
# dropping duplicated values
df1.drop_duplicates(inplace=True)
```

```python
df1.shape
```

# step 2 handling missing values

```python
df1.isna().sum().sort_values(ascending=False)[:6]
```

```python
df1[['company','agent']]=df1[['company','agent']].fillna(0)
```

```python
df1.head()
```

```python
hotel['company'].value_counts()
```

```python
hotel['company']
```

```python
hotel.describe()
```

```python
df1['children'].fillna(df1['children'].mean(),inplace=True)
```

```python
df1['country'].fillna('others',inplace=True)
```

```python
df1.isna().sum().sort_values(ascending=False)[:6]
```

```python
df1[df1['adults']+df1['babies']+df1['children']==0].shape
```

```python
df1.drop(df1[df1['adults']+df1['babies']+df1['children']==0].index,inplace=True)
```

```python
df1.head()
```

# step 3 to convert EDA into appropriate datatype

```python
#converting datatype of column 
df1[['children','company','agent']]=df1[['children','company','agent']].astype('int64')
```

```python
# changing datatype of column 
df1['reservation_status_date']=pd.to_datetime(df1['reservation_status_date'])
```

```python
df1['reservation_status_date']
```

# step4 : adding important columns


```python
# adding total staying days in hotels
df1['total_stay']=df1['stays_in_weekend_nights']+df1['stays_in_weekend_nights']
```

```python
# adding total people num as column ,i.e total people 
df1['total_people']=df1['adults']+df1['children']+df1['babies']
```
