# label based selection
with loc we need to specify the actule name of the column 

```python
# first row of the 
data.loc[0,'Name']
```

```python
# first 5 row
data.loc[:4,['Name','Sex','Age']]
```

# conditional selection

```python
data.loc[data['Age']==50,:]
```

```python
# rowa with age 50 and are female
# this is a sube t of the aboove dataframe by filtering out females
data.loc[(data['Age']==50)& (data['Sex']== 'female'),:]
```

```python
# row with age 50 or have fare greater than or equal to 200
data.loc[(data['Age']==50)| (data['Fare']>=200),:]
```

```python
# all the rows with null cabin column
```

```python
data.loc[data['Cabin'].isnull(),:]
```

```python
data['Embarked'].value_counts()
```

```python
# all thw row with c or q in embarked column
data.loc[data['Embarked'].isin(['C','Q']),:]
```

```python
data.loc[(data['Embarked']=='C') | (data['Embarked']=='Q'),:]
```
