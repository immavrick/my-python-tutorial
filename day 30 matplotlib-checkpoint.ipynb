```python
import numpy as np
import pandas as pd
```

```python
# let read the dataset
# using employee dataset
data=pd.read_csv('employee.csv')
#
data=data.loc[:199,:]
print(data.shape)

```

```python
data.head()
```

```python
# basic line chart plot
```

```python
plt.plot(data['Department'],data['Age'])
```

```python
# lets try to aggregrate the age usimg mean in each other 
x= data.pivot_table(index='Department',values='Age',aggfunc='mean')
x.plot(kind='line')
```

# multiple line plot

```python
x= data.pivot_table(index='Department',values='Age',aggfunc='mean')
x.plot(kind='line',color='r')
y= data.pivot_table(index='Department',values=['TotalWorkingYears'],aggfunc='mean')
y.plot(kind='line',color='b')
z= data.pivot_table(index='Department',values='HourlyRate',aggfunc='mean')
z.plot(kind='line',color='y')
w= data.pivot_table(index='Department',values='DailyRate',aggfunc='mean')
w.plot(kind='line',color='g')
plt.show
```

```python
#lets use different patterns for the lines

 

plt.rcParams['figure.figsize'] = (12, 5)

 

plt.subplot(221)

x = data.pivot_table(index = 'Department',values = ['Age'],aggfunc = 'mean')

plt.plot(x, color='red')

plt.ylabel('Age')

 

plt.subplot(222)

y = data.pivot_table(index = 'Department',values = ['TotalWorkingYears'],aggfunc = 'mean')

plt.plot(y, '-+', color = 'blue')

plt.ylabel('Experience')

 

plt.subplot(223)

z = data.pivot_table(index = 'Department',values = ['HourlyRate'],aggfunc = 'mean')

plt.plot(z, '--',  color ='green')

plt.ylabel('Salary')

 

plt.subplot(224)

a = data.pivot_table(index = 'Department',values = ['PerformanceRating'],aggfunc = 'mean')

plt.plot(a, '-*', color = 'black')

plt.ylabel('Performance')

 

plt.suptitle('Using Different Patterns for Different Line Plots')

plt.show()





```

# bar chart plot

```python
# analazing the attribute Department
```

```python
data['Department'].value_counts().plot(kind='bar',color='lightblue')
plt.ylabel('value')
plt.xlabel('Different Department')
plt.xticks(rotation=180)
plt.title('No. of employee in each Departments',fontsize=12)
plt.show()
```

```python
data.groupby('EducationField')['MonthlyRate'].mean()
```

```python
plt.rcParams['figure.figsize']=(10,4)
plt.bar(data['EducationField'],data['MonthlyRate'],width=0.7,align='center')
plt.title('Average Salary of Employee from Various Depatments')
plt.show()
```

```python
data.groupby(['Department','EducationField'])['MonthlyRate'].mean()
```

```python
plt.rcParams['figure.figsize']=(10,4)
x=data.pivot_table(index=['Department','EducationField'],values=['MonthlyRate'],aggfunc='mean')
x.plot(kind='bar',stacked=False)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.rcParams['figure.figsize']=(10,4)
x=data.pivot_table(index=['Department','EducationField'],values=['MonthlyRate'],aggfunc='mean')
x.plot(kind='bar',stacked=True)
plt.xticks(rotation=90)
plt.show()
```

```python
#crosstab
```

```python
x = pd.crosstab(index = data['Department'],

                columns = data['EducationField'],

                values = data['MonthlyRate'],

                aggfunc = 'mean')

x
```

```python
plt.rcParams['figure.figsize'] = (18, 6)

 

# keeping the Stacked option to **True**

x.plot(kind = 'bar', stacked = True)

plt.ylabel('Salary')

plt.xticks(rotation = 90)

plt.show()
```

```python
# setting the size of the plot

plt.rcParams['figure.figsize'] = (18, 5)

 

# keeping the stacked option to **false**

x.plot(kind = 'bar', stacked = False)

plt.xticks(rotation = 20)

plt.show()
```
