```python
plt.subplot()
```

```python
c=plt.figure(figsize=(18,10))
plt.subplot(3,1,1)
plt.plot(x,np.sin(x));

plt.subplot(3,1,2)
plt.plot(x,np.cos(x))

plt.subplot(3,1,3)
plt.plot(x,np.tan(x))
```
